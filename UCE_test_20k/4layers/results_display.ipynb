```python
cell_type_mapping = {
    "Ast": "Astrocytes",
    "End": "Endothelial_cells",
    "Exc": "Excitatory_neurons",
    "Fib": "Fibroblasts",
    "Inh": "Inhibitary_neurons",
    "Mic": "Microglia",
    "Oli": "Oligodendrocytes",
    "OPC": "OPCs",
    "Per": "Pericytes",
    "CAMs": "CAMs",
    "SMC": "SMC",
    "T cells": "T cells",
    "Epd": "Epd",
    "CPEC": "CPEC"
}

human_uce.obs['cell_type'] = human_uce.obs['major_cell_type'].map(cell_type_mapping)
```

```python
merged_adata = human_uce.concatenate(mouse_uce, 
                                    batch_key="species", 
                                    batch_categories=["human", "mouse"])
```
