```python
import scanpy as sc
```

```python
# Load the .h5ad file
frog = sc.read_h5ad('../data/frog.h5ad')
zebrafish = sc.read_h5ad('../data/zebrafish.h5ad')
```

```python
frog_genes = frog.var_names
zebrafish_genes = zebrafish.var_names
```

```python
print(frog_genes)
```

```python
print(zebrafish_genes)
```

```python
# Common Genes
common_genes = set(frog_genes).intersection(set(zebrafish_genes))
print(sorted(common_genes))  # optional: sort alphabetically
print(len(common_genes))
```

```python
# Differential Genes
frog_unique = frog_genes.difference(zebrafish_genes)
print(len(frog_unique))

zebrafish_unique = zebrafish_genes.difference(frog_genes)
print(len(zebrafish_unique))
```
