```python
import scanpy as sc
import os
import pandas as pd
import numpy as np
import anndata as ad
import seaborn as sns
from matplotlib import pyplot as plt
from glob import glob
```

```python
human_adata = sc.read_h5ad("data/human_subsetted.h5ad")
```

```python
sample_list = ["data/SNS2-2.h5ad", "data/SNS3_1.h5ad", "data/SNS3_2.h5ad", "data/SNS3_4.h5ad"]
```

```python
mouse_adatas = [sc.read_h5ad(path) for path in sample_list]  
```

```python
mouse_adata = ad.concat(mouse_adatas, join="outer", label="batch", keys=sample_list)
```

```python
human_adata.obs.rename(columns={"major_cell_type":"cell_type"}, inplace=True)
human_adata.obs.rename(columns={"projid":"sample"}, inplace=True)
mouse_adata.obs.rename(columns={"orig.ident":"sample"}, inplace=True)
```

```python
mouse_adata.write("data/mouse.h5ad")
human_adata.write("data/human.h5ad")
```
